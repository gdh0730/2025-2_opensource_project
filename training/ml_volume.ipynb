```python
import os
from pathlib import Path
import unicodedata as ud
import re
import gc

import numpy as np
import pandas as pd

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler
from torch.optim.lr_scheduler import StepLR
from tqdm.auto import tqdm
import matplotlib.pyplot as plt

# =========================
# 경로 & 하이퍼파라미터
# =========================
BASE_DIR = Path('/content/drive/MyDrive/metrovision-ai/LSTM/vol')  # 필요시 수정
DATA_DIR = BASE_DIR / 'data'
PARQUET_DIR = BASE_DIR / 'parquet'
PARQUET_DIR.mkdir(parents=True, exist_ok=True)
MODEL_DIR = BASE_DIR / 'models'
MODEL_DIR.mkdir(parents=True, exist_ok=True)

VOL_PARQUET = PARQUET_DIR / "volume_clean_all_scaled.parquet"
CKPT_PATH = MODEL_DIR / 'volume_seq2seq_gcn_lstm_scaled.pt'

# 시계열 윈도우 설정
T = 24   # 입력: 최근 24시간
H = 6    # 출력: 향후 6시간

# 다운사이징 옵션
LINK_TOP_K = 6   # 대표 도로명 상위 10개 사용 (실제론 아래 top_roads 리스트 사용)
STRIDE = 6        # 슬라이딩 윈도우 stride (6시간 간격 샘플링)

# ===== 모델 용량 관련 하이퍼파라미터 =====
GCN_HIDDEN_DIM = 64       # GCN hidden
GCN_OUT_DIM    = 64       # GCN 출력 (링크 embedding 역할)

ENC_HIDDEN_DIM = 128      # Encoder LSTM hidden
ENC_LAYERS     = 2        # Encoder LSTM layer 수

FC_HIDDEN_DIM  = 128      # Decoder FC hidden

DROPOUT_P      = 0.0      # 필요시 0.1~0.2로 조정

# 학습 하이퍼파라미터
BATCH_SIZE = 1024
LR = 1e-3
EPOCHS = 30

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("device:", device)

# ===============================================
# [교통량 Vol 모델] 전역 설정
# ===============================================

VOL_ID_COLS = [
    'spot_dir_id', '지점번호', '지점명', '방향', '도로명'
]

VOL_TIME_COLS = [
    '일자', 'timestamp', 'hour', 'day_of_week', 'is_weekend', 'month',
]

VOL_NUMERIC_COLS = [
    'vol',             # 타겟
    'spot_mean_vol',   # 정적 피처
    'spot_max_vol',    # 정적 피처
    'lag_1h', 'lag_24h',
    'roll_mean_3h', 'roll_std_3h',
    'roll_mean_6h', 'roll_std_6h',
]

VOL_CATEGORICAL_COLS = [
    '요일', '구분', '방향', '도로명', '지점명',
]

VOL_NUMERIC_RULES = {
    'vol': dict(min=0, max=None, drop_if_invalid=True),
    'spot_mean_vol': dict(min=0, max=None, drop_if_invalid=True),
    'spot_max_vol': dict(min=0, max=None, drop_if_invalid=True),
}

# speed 모델에서 쓰던 대표 도로 예시 그대로 사용
top_roads = ['남부순환로', '동일로', '천호대로', '강변북로', '올림픽대로', '양재대로']


# ===============================================
# 1. 파일명에서 연/월 파싱
# ===============================================
def parse_year_month_vol(norm_name: str):
    """
    교통량 엑셀 파일명에서 (연,월) 추출.
    예) '01월 서울시 교통량 조사자료(2025).xlsx' -> (2025, 1)
        '2025년 1월 서울시 교통량 조사자료.xlsx' -> (2025, 1)
    """
    m1 = re.search(r'(\d{4})년\s*(\d{1,2})월', norm_name)
    if m1:
        year = int(m1.group(1))
        month = int(m1.group(2))
        return (year, month)

    m2 = re.search(r'(\d{1,2})월.*\((\d{4})\)', norm_name)
    if m2:
        month = int(m2.group(1))
        year = int(m2.group(2))
        return (year, month)

    return (9999, 99)


# ===============================================
# 2. 지점명에서 도로명 추출
# ===============================================
def extract_road_from_spot_name(spot_name: str) -> str:
    """
    예: '성산로(금화터널)' -> '성산로'
    괄호가 없으면 전체 문자열 사용.
    """
    if pd.isna(spot_name):
        return ""
    s = str(spot_name)
    m = re.match(r'([^()]+)', s)
    if m:
        return m.group(1).strip()
    return s.strip()


# ===============================================
# 3. 파일 단위 wide -> long 변환 (메모리 최적화)
# ===============================================
def load_volume_long(top_roads=None):
    """
    data 폴더 안의 '교통량' 엑셀 파일들을
      - 파일 하나씩 읽어서 도로명 추출
      - 대표 도로 필터링
      - wide -> long 변환
    후 concat
    """
    files = []
    print("[VOL] Found raw files:")
    for f in DATA_DIR.glob("*.xlsx"):
        norm_name = ud.normalize('NFC', f.name)
        if "교통량" in norm_name:
            ym = parse_year_month_vol(norm_name)
            files.append((f, norm_name, ym))

    files = sorted(files, key=lambda x: x[2])
    if len(files) == 0:
        raise FileNotFoundError("DATA_DIR에 '교통량' 엑셀 파일이 없습니다.")

    for f, norm_name, ym in files:
        print("  ", norm_name, "->", f, "year-month:", ym)

    long_list = []

    for f, norm_name, ym in files:
        print("[VOL] loading (wide):", f)
        df = pd.read_excel(f, sheet_name=1)

        # 도로명 추출
        df['도로명'] = df['지점명'].apply(extract_road_from_spot_name)

        # 대표 도로 필터링
        if top_roads is not None:
            df = df[df['도로명'].isin(top_roads)].copy()
            if df.empty:
                print(f"[VOL]   -> top_roads 필터 후 남는 행 없음, skip")
                continue

        # 시간 컬럼(0시~23시)
        hour_cols = [c for c in df.columns if c.endswith('시')]
        print(f"[VOL]   hour_cols ({len(hour_cols)}개):", hour_cols[:6], "...")

        df['일자'] = pd.to_datetime(df['일자'].astype(str), format='%Y%m%d', errors='coerce')

        id_cols = [
            '일자', '요일', '지점명', '지점번호', '방향', '구분', '도로명'
        ]

        long_df = df.melt(
            id_vars=id_cols,
            value_vars=hour_cols,
            var_name='시간대',
            value_name='vol'
        )

        long_df['hour'] = long_df['시간대'].str.extract(r'(\d+)').astype(int)
        long_df['timestamp'] = long_df['일자'] + pd.to_timedelta(long_df['hour'], unit='h')

        long_df['spot_dir_id'] = (
            long_df['지점번호'].astype(str).str.strip() + "_" +
            long_df['방향'].astype(str).str.strip()
        )

        long_list.append(long_df)

    if not long_list:
        raise RuntimeError("[VOL] long 데이터가 아무 것도 생성되지 않았습니다. top_roads 필터 조건을 확인하세요.")

    vol_long = pd.concat(long_list, ignore_index=True)
    vol_long = vol_long.sort_values(['spot_dir_id', 'timestamp']).reset_index(drop=True)

    print("[VOL] 최종 long_df 행 수:", len(vol_long))
    return vol_long


# ===============================================
# 4. 기본 시계열 피처 추가
# ===============================================
def add_basic_features_volume(long_df: pd.DataFrame) -> pd.DataFrame:
    df = long_df.copy()
    df['day_of_week'] = df['timestamp'].dt.dayofweek.astype('int32')
    df['is_weekend'] = (df['day_of_week'] >= 5).astype(int)
    df['month'] = df['timestamp'].dt.month.astype('int32')

    df['sin_hour'] = np.sin(2 * np.pi * df['hour'] / 24)
    df['cos_hour'] = np.cos(2 * np.pi * df['hour'] / 24)
    return df


# ===============================================
# 5. 컬럼 점검
# ===============================================
def inspect_volume_columns(df: pd.DataFrame, n_top: int = 10):
    print("[VOL] ===== 전체 행 수 =====")
    print(len(df))
    print()

    print("[VOL] ===== 컬럼별 타입 =====")
    print(df.dtypes)
    print()

    print("[VOL] ===== 컬럼별 NaN 개수 =====")
    nan_counts = df.isna().sum().sort_values(ascending=False)
    print(nan_counts)
    print()

    numeric_cols_present = [c for c in VOL_NUMERIC_COLS if c in df.columns]
    if numeric_cols_present:
        print("[VOL] ===== 수치형 컬럼 통계 =====")
        print(df[numeric_cols_present].describe())
        print()

        print("[VOL] ===== 수치형 컬럼별 규칙 기반 이상치 개수 =====")
        for col in numeric_cols_present:
            rule = VOL_NUMERIC_RULES.get(col)
            if not rule:
                continue
            s = df[col]
            n_nan = s.isna().sum()
            conds = []
            if rule.get('min') is not None:
                conds.append(s < rule['min'])
            if rule.get('max') is not None:
                conds.append(s > rule['max'])
            if conds:
                invalid = conds[0]
                for c in conds[1:]:
                    invalid |= c
                n_invalid = invalid.sum()
            else:
                n_invalid = 0
            print(f"[VOL] {col}: NaN={n_nan}, invalid_range={n_invalid}")
        print()

    cat_cols_present = [c for c in VOL_CATEGORICAL_COLS if c in df.columns]
    if cat_cols_present:
        print("[VOL] ===== 범주형 컬럼 값 분포 =====")
        for col in cat_cols_present:
            print(f"[{col}]")
            print(df[col].value_counts(dropna=False).head(n_top))
            print()

    id_cols_present = [c for c in VOL_ID_COLS if c in df.columns]
    if id_cols_present:
        print("[VOL] ===== ID 컬럼 NaN/공백 개수 =====")
        for col in id_cols_present:
            s = df[col]
            n_nan = s.isna().sum()
            n_blank = (s.astype(str).str.strip() == '').sum()
            print(f"{col}: NaN={n_nan}, blank={n_blank}")


# ===============================================
# 6. 전처리
# ===============================================
def clean_volume_all(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    print("[VOL] === [1] inf → NaN 처리 ===")
    df.replace([np.inf, -np.inf], np.nan, inplace=True)

    print("[VOL] === [2] 타입 정규화 ===")
    if 'timestamp' in df.columns:
        df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
    if '일자' in df.columns:
        if not np.issubdtype(df['일자'].dtype, np.datetime64):
            df['일자'] = pd.to_datetime(df['일자'], errors='coerce')

    for col in VOL_NUMERIC_COLS:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')

    for col in VOL_CATEGORICAL_COLS:
        if col in df.columns:
            df[col] = df[col].astype(str).str.strip()

    for col in VOL_ID_COLS:
        if col in df.columns:
            df[col] = df[col].astype(str).str.strip()

    print("[VOL] === [3] 필수 ID/시간/타겟 정제 ===")
    before = len(df)
    df = df[~df['spot_dir_id'].isna()]
    df = df[df['spot_dir_id'].astype(str).str.strip() != '']
    df = df[~df['timestamp'].isna()]
    df = df[~df['vol'].isna()]
    after = len(df)
    print(f"[VOL] 필수 기준: {before} -> {after} 행")

    print("[VOL] === [4] 수치형 컬럼 범위 규칙 적용 ===")
    before = len(df)
    for col, rule in VOL_NUMERIC_RULES.items():
        if col not in df.columns:
            continue
        s = df[col]
        cond_keep = pd.Series(True, index=df.index)
        if rule.get('min') is not None:
            cond_keep &= (s >= rule['min']) | s.isna()
        if rule.get('max') is not None:
            cond_keep &= (s <= rule['max']) | s.isna()
        if rule.get('drop_if_invalid', False):
            cond_keep &= ~s.isna()
        df = df[cond_keep]
    after2 = len(df)
    print(f"[VOL] 수치형 범위/NaN 정제: {before} -> {after2} 행")

    print("[VOL] === [5] 범주형 NaN/공백 정리 ===")
    for col in VOL_CATEGORICAL_COLS:
        if col not in df.columns:
            continue
        s = df[col]
        s = s.fillna('미상')
        s = s.replace({'': '미상', 'nan': '미상', 'NaN': '미상'})
        df[col] = s

    print("[VOL] === [6] spot_dir_id 기준 정적 통계 재계산 ===")
    spot_stats = df.groupby('spot_dir_id')['vol'].agg(['mean', 'max']).reset_index()
    spot_stats = spot_stats.rename(columns={'mean': 'spot_mean_vol', 'max': 'spot_max_vol'})
    df = df.drop(columns=['spot_mean_vol', 'spot_max_vol'], errors='ignore')
    df = df.merge(spot_stats, on='spot_dir_id', how='left')

    print("[VOL] === [7] 최종 NaN/이상치 점검 ===")
    inspect_volume_columns(df)

    df = df.sort_values(['spot_dir_id', 'timestamp']).reset_index(drop=True)
    return df


# ===============================================
# 7. Rolling / Lag Feature
# ===============================================
def add_rolling_features_volume(df: pd.DataFrame) -> pd.DataFrame:
    df = df.sort_values(['spot_dir_id', 'timestamp'])
    g = df.groupby('spot_dir_id')

    df['lag_1h']  = g['vol'].shift(1)
    df['lag_24h'] = g['vol'].shift(24)

    df['roll_mean_3h'] = g['vol'].rolling(window=3, min_periods=1).mean().reset_index(level=0, drop=True)
    df['roll_std_3h']  = g['vol'].rolling(window=3, min_periods=1).std().reset_index(level=0, drop=True)

    df['roll_mean_6h'] = g['vol'].rolling(window=6, min_periods=1).mean().reset_index(level=0, drop=True)
    df['roll_std_6h']  = g['vol'].rolling(window=6, min_periods=1).std().reset_index(level=0, drop=True)

    for col in ['roll_std_3h', 'roll_std_6h']:
        df[col] = df[col].fillna(0.0)

    df = df.dropna(subset=['lag_1h', 'lag_24h'])
    df = df.sort_values(['spot_dir_id', 'timestamp']).reset_index(drop=True)
    return df


# ===============================================
# 8. Dataset
# ===============================================
class VolumeWindowDataset(Dataset):
    def __init__(
        self,
        df: pd.DataFrame,
        T: int = 24,
        H: int = 6,
        mode: str = 'train',
        spot_id_to_idx=None,
        scaler_dynamic: StandardScaler | None = None,
        scaler_static: StandardScaler | None = None,
        stride: int = 1,
    ):
        self.T = T
        self.H = H
        self.stride = stride
        self.mode = mode

        self.dynamic_cols = [
            'vol', 'hour', 'day_of_week', 'is_weekend', 'sin_hour', 'cos_hour',
            'lag_1h', 'lag_24h',
            'roll_mean_3h', 'roll_std_3h',
            'roll_mean_6h', 'roll_std_6h',
        ]
        self.static_num_cols = ['spot_mean_vol', 'spot_max_vol']
        self.future_time_cols = ['hour', 'day_of_week', 'is_weekend', 'sin_hour', 'cos_hour']

        df = df.sort_values(['spot_dir_id', 'timestamp'])

        if mode == 'train':
            spot_ids = sorted(df['spot_dir_id'].unique().tolist())
            self.spot_id_to_idx = {sid: i for i, sid in enumerate(spot_ids)}
        else:
            assert spot_id_to_idx is not None
            self.spot_id_to_idx = spot_id_to_idx

        self.num_spots = len(self.spot_id_to_idx)

        if mode == 'train':
            self.scaler_dynamic = StandardScaler()
            self.scaler_static  = StandardScaler()
        else:
            assert scaler_dynamic is not None and scaler_static is not None
            self.scaler_dynamic = scaler_dynamic
            self.scaler_static  = scaler_static

        X_dyn_list, X_stat_list = [], []
        spot_idx_list, y_list, last_vol_list = [], [], []
        future_time_list = []

        for spot_id, g in df.groupby('spot_dir_id'):
            if spot_id not in self.spot_id_to_idx:
                continue
            if g['vol'].isna().any():
                continue
            if len(g) < (T + H + 7*24):
                continue

            g = g.reset_index(drop=True)

            dyn = g[self.dynamic_cols].values.astype(float)
            stat = g[self.static_num_cols].iloc[0].values.astype(float)
            target_series = g['vol'].values.astype(float)

            N = len(g) - T - H + 1
            for i in range(0, N, self.stride):
                X_dyn_list.append(dyn[i:i+T])
                X_stat_list.append(stat)

                y_seq = target_series[i+T:i+T+H]
                y_list.append(y_seq)

                last_vol_list.append(target_series[i+T-1])

                spot_idx_list.append(self.spot_id_to_idx[spot_id])

                future_slice = g.iloc[i+T:i+T+H]
                future_time = future_slice[self.future_time_cols].values.astype(float)
                future_time_list.append(future_time)

        X_dyn = np.stack(X_dyn_list)          # (N, T, dyn_dim)
        X_stat = np.stack(X_stat_list)        # (N, static_dim)
        y = np.stack(y_list)                  # (N, H)
        last_vol = np.array(last_vol_list)    # (N,)
        spot_idx = np.array(spot_idx_list)    # (N,)
        future_time = np.stack(future_time_list)  # (N, H, time_dim)

        print(f"[VOL-{mode}] 생성된 윈도우 샘플 수:", X_dyn.shape[0])

        N, T_, D = X_dyn.shape

        if mode == 'train':
            X_dyn_2d = X_dyn.reshape(-1, D)
            X_dyn_scaled_2d = self.scaler_dynamic.fit_transform(X_dyn_2d)
            X_dyn_scaled = X_dyn_scaled_2d.reshape(N, T_, D)
            X_stat_scaled = self.scaler_static.fit_transform(X_stat)
        else:
            X_dyn_2d = X_dyn.reshape(-1, D)
            X_dyn_scaled_2d = self.scaler_dynamic.transform(X_dyn_2d)
            X_dyn_scaled = X_dyn_scaled_2d.reshape(N, T_, D)
            X_stat_scaled = self.scaler_static.transform(X_stat)

        # 타깃 스케일 (dynamic_cols[0] == 'vol')
        self.vol_mean = float(self.scaler_dynamic.mean_[0])
        self.vol_std  = float(self.scaler_dynamic.scale_[0] + 1e-8)

        y_scaled = (y - self.vol_mean) / self.vol_std
        last_vol_scaled = (last_vol - self.vol_mean) / self.vol_std

        self.X_dyn = torch.tensor(X_dyn_scaled, dtype=torch.float32)
        self.X_stat = torch.tensor(X_stat_scaled, dtype=torch.float32)
        self.y = torch.tensor(y_scaled, dtype=torch.float32)
        self.last_vol = torch.tensor(last_vol_scaled, dtype=torch.float32)
        self.spot_idx = torch.tensor(spot_idx, dtype=torch.long)
        self.future_time = torch.tensor(future_time, dtype=torch.float32)

    def __len__(self):
        return self.X_dyn.shape[0]

    def __getitem__(self, idx):
        return (
            self.X_dyn[idx],       # (T, dyn_dim)
            self.spot_idx[idx],    # ()
            self.X_stat[idx],      # (static_dim,)
            self.y[idx],           # (H,)
            self.last_vol[idx],    # ()
            self.future_time[idx], # (H, time_dim)
        )


# ===============================================
# 9. Spot 그래프 생성
# ===============================================
def build_spot_graph(df: pd.DataFrame, spot_id_to_idx: dict):
    """
    노드: spot_dir_id
    노드 피처: [spot_mean_vol, spot_max_vol] (z-score 스케일링)
    인접행렬: 자기 자신 + 같은 도로명
    """
    n = len(spot_id_to_idx)
    print("[VOL] spot 개수 (그래프 노드 수):", n)

    stats = df.groupby('spot_dir_id').agg({
        '도로명': 'first',
        'spot_mean_vol': 'first',
        'spot_max_vol': 'first'
    }).reset_index()

    global_mean = stats['spot_mean_vol'].mean()
    global_max  = stats['spot_max_vol'].mean()

    raw_feat = []
    for _, row in stats.iterrows():
        mv = row['spot_mean_vol'] if not np.isnan(row['spot_mean_vol']) else global_mean
        xv = row['spot_max_vol']  if not np.isnan(row['spot_max_vol'])  else global_max
        raw_feat.append([mv, xv])
    raw_feat = np.array(raw_feat, dtype=np.float32)

    feat_mean = raw_feat.mean(axis=0)
    feat_std  = raw_feat.std(axis=0) + 1e-6
    norm_feat = (raw_feat - feat_mean) / feat_std

    X = np.zeros((n, 2), dtype=np.float32)
    road_name_list = [''] * n

    for i, row in stats.iterrows():
        sid = row['spot_dir_id']
        if sid not in spot_id_to_idx:
            continue
        idx = spot_id_to_idx[sid]
        X[idx] = norm_feat[i]
        road_name_list[idx] = row['도로명']

    A = np.zeros((n, n), dtype=np.float32)
    for i in range(n):
        for j in range(n):
            if i == j or road_name_list[i] == road_name_list[j]:
                A[i, j] = 1.0

    D = np.diag(A.sum(axis=1) + 1e-8)
    D_inv_sqrt = np.linalg.inv(np.sqrt(D))
    A_hat = D_inv_sqrt @ A @ D_inv_sqrt

    X_tensor = torch.from_numpy(X).float()
    A_hat_tensor = torch.from_numpy(A_hat).float()
    return X_tensor, A_hat_tensor


# ===============================================
# 10. 모델 정의 (GCN + LSTM Seq2Seq)
# ===============================================
class RoadGCN(nn.Module):
    def __init__(self, in_dim: int, hidden_dim: int, out_dim: int,
                 X0: torch.Tensor, A_hat: torch.Tensor):
        super().__init__()
        self.register_buffer("A_hat", A_hat)
        self.register_buffer("X0", X0)
        self.fc1 = nn.Linear(in_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, out_dim)

    def forward(self):
        H1 = torch.relu(self.A_hat @ self.fc1(self.X0))
        Z  = torch.relu(self.A_hat @ self.fc2(H1))
        return Z  # (N, out_dim)


class EncoderGCN(nn.Module):
    def __init__(
        self,
        dynamic_dim: int,
        gcn_in_dim: int,
        gcn_hidden_dim: int,
        gcn_out_dim: int,
        X0: torch.Tensor,
        A_hat: torch.Tensor,
        lstm_hidden_dim: int = 64,
        lstm_layers: int = 1,
        dropout_p: float = 0.0,
    ):
        super().__init__()
        self.gcn = RoadGCN(gcn_in_dim, gcn_hidden_dim, gcn_out_dim, X0, A_hat)
        self.gcn_out_dim = gcn_out_dim
        self.lstm_input_dim = dynamic_dim + gcn_out_dim
        self.lstm = nn.LSTM(
            input_size=self.lstm_input_dim,
            hidden_size=lstm_hidden_dim,
            num_layers=lstm_layers,
            batch_first=True,
            dropout=dropout_p if lstm_layers > 1 else 0.0,
        )

    def forward(self, x_dyn, link_idx):
        Z = self.gcn()                  # (N, gcn_out_dim)
        e_link = Z[link_idx]            # (B, gcn_out_dim)

        B, T, _ = x_dyn.shape
        e_expanded = e_link.unsqueeze(1).repeat(1, T, 1)

        lstm_input = torch.cat([x_dyn, e_expanded], dim=-1)
        outputs, (h_n, c_n) = self.lstm(lstm_input)
        return h_n, c_n, e_link


class DecoderGCN(nn.Module):
    def __init__(
        self,
        gcn_out_dim: int,
        static_dim: int,
        lstm_hidden_dim: int,
        horizon: int,
        time_dim: int,
        fc_hidden_dim: int = 128,
        lstm_layers: int = 1,
        dropout_p: float = 0.0,
        constrain_nonneg: bool = True,  # ✅ raw(vol) >= 0 제약 on/off
        clamp_min: float = 0.0,          # ✅ raw 공간 clamp 최소값(기본 0)
    ):
        super().__init__()

        self.input_dim = 1 + gcn_out_dim + static_dim + time_dim
        self.lstm = nn.LSTM(
            input_size=self.input_dim,
            hidden_size=lstm_hidden_dim,
            num_layers=lstm_layers,
            batch_first=True,
        )

        self.dropout = nn.Dropout(dropout_p)

        # ✅ Softplus 제거: scaled 공간에 제약을 걸면 분포가 깨짐
        self.fc = nn.Sequential(
            nn.Linear(lstm_hidden_dim + static_dim + gcn_out_dim, fc_hidden_dim),
            nn.ReLU(),
            nn.Linear(fc_hidden_dim, 1),
        )

        self.horizon = horizon

        # ✅ 제약 관련 옵션
        self.constrain_nonneg = constrain_nonneg
        self.clamp_min = float(clamp_min)

        # ✅ 타깃(vol) 스케일러를 "모델 내부"에 저장 (state_dict에 포함됨)
        #    (학습 시작 직후 vol_train_ds.vol_mean/std를 주입해야 함)
        self.register_buffer("target_mean", torch.tensor(0.0, dtype=torch.float32))
        self.register_buffer("target_std",  torch.tensor(1.0, dtype=torch.float32))

    @torch.no_grad()
    def set_target_scaler(self, mean: float, std: float):
        """
        vol_train_ds.vol_mean / vol_train_ds.vol_std (또는 ckpt['target_mean/std'])
        를 디코더에 주입. raw 공간 제약을 위해 필수.
        """
        self.target_mean.fill_(float(mean))
        self.target_std.fill_(float(std))

    def _apply_raw_nonneg_constraint(self, y_scaled: torch.Tensor) -> torch.Tensor:
        """
        y_scaled: (B,) 또는 (B, H)  (scaled 공간)
        -> raw로 복원 후 clamp -> 다시 scaled로 반환
        """
        if not self.constrain_nonneg:
            return y_scaled

        # scaled -> raw
        raw = y_scaled * self.target_std + self.target_mean

        # raw에서만 제약 (교통량/유량은 0 이상)
        raw = torch.clamp(raw, min=self.clamp_min)

        # raw -> scaled
        y_scaled2 = (raw - self.target_mean) / (self.target_std + 1e-8)
        return y_scaled2

    def forward(
        self,
        h0, c0,
        e_link,
        x_static,
        future_time,
        y=None,
        last_speed=None,
        teacher_forcing: bool = False
    ):
        """
        입력/출력은 기존 코드와 동일 (scaled 공간)
        - last_speed: scaled (B,)
        - y(teacher forcing): scaled (B,H)
        - return outputs: scaled (B,H)
        """
        B = e_link.size(0)
        H = self.horizon
        outputs = []

        # prev_speed는 scaled 공간에서 굴림
        if last_speed is None:
            raise ValueError("DecoderGCN.forward(): last_speed (scaled)가 필요합니다.")
        prev_speed = last_speed.view(B, 1)

        h, c = h0, c0

        for t in range(H):
            time_feat_t = future_time[:, t, :]  # (B, time_dim)

            dec_in = torch.cat(
                [prev_speed, e_link, x_static, time_feat_t],
                dim=-1
            )  # (B, input_dim)
            dec_in = dec_in.unsqueeze(1)  # (B, 1, input_dim)

            out, (h, c) = self.lstm(dec_in, (h, c))
            h_t = out[:, -1, :]  # (B, hidden)

            fc_in = torch.cat([h_t, x_static, e_link], dim=-1)
            fc_in = self.dropout(fc_in)

            # 1) scaled 예측 (제약 전)
            step_pred = self.fc(fc_in).squeeze(-1)  # (B,)

            # 2) ✅ raw 공간에서 non-negative 제약 적용 후 다시 scaled로 복귀
            step_pred = self._apply_raw_nonneg_constraint(step_pred)

            outputs.append(step_pred.unsqueeze(1))

            # 다음 스텝 입력(prev_speed) 결정
            if teacher_forcing and (y is not None):
                # teacher forcing은 데이터(y_scaled) 그대로 사용
                prev_speed = y[:, t].view(B, 1)
            else:
                # 예측값을 다음 입력으로 사용 (detach 유지)
                prev_speed = step_pred.detach().view(B, 1)

        outputs = torch.cat(outputs, dim=1)  # (B, H) scaled
        return outputs


class Seq2SeqGCNLSTM(nn.Module):
    def __init__(
        self,
        dynamic_dim: int,
        static_dim: int,
        horizon: int,
        time_dim: int,
        gcn_in_dim: int,
        gcn_hidden_dim: int,
        gcn_out_dim: int,
        X0: torch.Tensor,
        A_hat: torch.Tensor,
        lstm_hidden_dim: int = 64,
        lstm_layers: int = 1,
        fc_hidden_dim: int = 128,
        dropout_p: float = 0.0,
    ):
        super().__init__()
        self.encoder = EncoderGCN(
            dynamic_dim=dynamic_dim,
            gcn_in_dim=gcn_in_dim,
            gcn_hidden_dim=gcn_hidden_dim,
            gcn_out_dim=gcn_out_dim,
            X0=X0,
            A_hat=A_hat,
            lstm_hidden_dim=lstm_hidden_dim,
            lstm_layers=lstm_layers,
            dropout_p=dropout_p,
        )
        self.decoder = DecoderGCN(
            gcn_out_dim=gcn_out_dim,
            static_dim=static_dim,
            lstm_hidden_dim=lstm_hidden_dim,
            horizon=horizon,
            time_dim=time_dim,
            fc_hidden_dim=fc_hidden_dim,
            lstm_layers=lstm_layers,
            dropout_p=dropout_p,
        )

    def forward(
        self,
        x_dyn,
        link_idx,
        x_static,
        future_time,
        y=None,
        last_speed=None,
        teacher_forcing: bool = False
    ):
        h0, c0, e_link = self.encoder(x_dyn, link_idx)
        y_hat = self.decoder(
            h0, c0,
            e_link,
            x_static,
            future_time=future_time,
            y=y,
            last_speed=last_speed,
            teacher_forcing=teacher_forcing
        )
        return y_hat


# ===============================================
# 11. 평가 함수들
# ===============================================
def compute_regression_metrics(y_true: np.ndarray, y_pred: np.ndarray):
    y_true = y_true.astype(float)
    y_pred = y_pred.astype(float)
    diff = y_true - y_pred

    mse = np.mean(diff ** 2)
    rmse = np.sqrt(mse)
    mae = np.mean(np.abs(diff))
    denom = np.abs(y_true) + 1e-6
    mape = np.mean(np.abs(diff) / denom)
    var = np.var(y_true)
    r2 = 1.0 - mse / (var + 1e-12)

    return {
        "mse": mse,
        "rmse": rmse,
        "mae": mae,
        "mape": mape,
        "r2": r2,
    }


def print_metrics(name: str, metrics: dict):
    print(f"\n[{name}] Overall Metrics")
    print(f"  MSE  : {metrics['mse']:.4f}")
    print(f"  RMSE : {metrics['rmse']:.4f}")
    print(f"  MAE  : {metrics['mae']:.4f}")
    print(f"  MAPE : {metrics['mape']*100:.2f}%")
    print(f"  R^2  : {metrics['r2']:.4f}  (1에 가까울수록 좋음)")


def evaluate_volume_model_test(model, test_loader, device, vol_mean, vol_std, horizon: int):
    model.eval()
    all_true = []
    all_pred = []

    test_bar = tqdm(test_loader, desc="[VOL] Test (Model)", leave=False)

    with torch.no_grad():
        for x_dyn, spot_idx, x_static, y_scaled, last_vol_scaled, future_time in test_bar:
            x_dyn = x_dyn.to(device)
            spot_idx = spot_idx.to(device)
            x_static = x_static.to(device)
            y_scaled = y_scaled.to(device)
            last_vol_scaled = last_vol_scaled.to(device)
            future_time = future_time.to(device)

            y_hat_scaled = model(
                x_dyn, spot_idx, x_static,
                future_time=future_time,
                y=None,
                last_speed=last_vol_scaled,
                teacher_forcing=False
            )

            y_hat = y_hat_scaled * vol_std + vol_mean
            y = y_scaled * vol_std + vol_mean

            all_true.append(y.cpu().numpy())
            all_pred.append(y_hat.cpu().numpy())

    y_true = np.concatenate(all_true, axis=0)
    y_pred = np.concatenate(all_pred, axis=0)

    overall_metrics = compute_regression_metrics(y_true, y_pred)
    print_metrics("VOL Model", overall_metrics)

    print("\n[VOL Model] Horizon-wise Metrics")
    for h in range(horizon):
        m = compute_regression_metrics(y_true[:, h], y_pred[:, h])
        print(f"  H+{h+1}: RMSE={m['rmse']:.3f}, MAE={m['mae']:.3f}, "
              f"MAPE={m['mape']*100:.2f}%, R^2={m['r2']:.4f}")

    return overall_metrics, (y_true, y_pred)


def eval_naive_last_value_volume(test_loader, vol_mean, vol_std, horizon: int):
    all_true = []
    all_pred = []

    for x_dyn, spot_idx, x_static, y_scaled, last_vol_scaled, future_time in test_loader:
        y = (y_scaled.numpy() * vol_std) + vol_mean

        last_scaled = x_dyn[:, -1, 0].numpy()
        last_raw = last_scaled * vol_std + vol_mean

        y_hat = np.tile(last_raw.reshape(-1, 1), (1, horizon))

        all_true.append(y)
        all_pred.append(y_hat)

    y_true = np.concatenate(all_true, axis=0)
    y_pred = np.concatenate(all_pred, axis=0)

    overall_metrics = compute_regression_metrics(y_true, y_pred)
    print_metrics("VOL Naive-LastValue", overall_metrics)

    print("\n[VOL Naive-LastValue] Horizon-wise Metrics")
    for h in range(horizon):
        m = compute_regression_metrics(y_true[:, h], y_pred[:, h])
        print(f"  H+{h+1}: RMSE={m['rmse']:.3f}, MAE={m['mae']:.3f}, "
              f"MAPE={m['mape']*100:.2f}%, R^2={m['r2']:.4f}")

    return overall_metrics


def eval_naive_last_24h_volume(test_loader, vol_mean, vol_std,
                               horizon: int, lag24_idx: int, lag24_mean: float, lag24_std: float):
    all_true = []
    all_pred = []

    for x_dyn, spot_idx, x_static, y_scaled, last_vol_scaled, future_time in test_loader:
        y = (y_scaled.numpy() * vol_std) + vol_mean

        lag24_scaled = x_dyn[:, -1, lag24_idx].numpy()
        lag24_raw = lag24_scaled * lag24_std + lag24_mean

        y_hat = np.tile(lag24_raw.reshape(-1, 1), (1, horizon))

        all_true.append(y)
        all_pred.append(y_hat)

    y_true = np.concatenate(all_true, axis=0)
    y_pred = np.concatenate(all_pred, axis=0)

    overall_metrics = compute_regression_metrics(y_true, y_pred)
    print_metrics("VOL Naive-Last24h", overall_metrics)

    print("\n[VOL Naive-Last24h] Horizon-wise Metrics")
    for h in range(horizon):
        m = compute_regression_metrics(y_true[:, h], y_pred[:, h])
        print(f"  H+{h+1}: RMSE={m['rmse']:.3f}, MAE={m['mae']:.3f}, "
              f"MAPE={m['mape']*100:.2f}%, R^2={m['r2']:.4f}")

    return overall_metrics


# ===============================================
# 12. 학습 엔트리 포인트
# ===============================================
def main():
    # 1) parquet 준비 (이미 있다면 생략 가능)
    if not VOL_PARQUET.exists():
        # 1) 데이터 로딩 & 전처리
        print("[VOL] speed 모델에서 정의된 top_roads 사용:", top_roads)
        vol_long = load_volume_long(top_roads=top_roads)

        print("[VOL] 대표 도로 필터 적용 후 long 행 수:", len(vol_long))
        vol_long = add_basic_features_volume(vol_long)

        vol_clean = clean_volume_all(vol_long)
        del vol_long
        gc.collect()

        vol_clean = add_rolling_features_volume(vol_clean)

        vol_clean.to_parquet(VOL_PARQUET, index=False)
        print("[VOL] 저장:", VOL_PARQUET)

    vol_clean = pd.read_parquet(VOL_PARQUET)
    vol_clean = vol_clean.sort_values(['spot_dir_id', 'timestamp']).reset_index(drop=True)

    print("[VOL] 최종 vol_clean 행 수:", len(vol_clean))
    print("[VOL] 기간:", vol_clean['timestamp'].min(), "~", vol_clean['timestamp'].max())
    print("[VOL] 지점 수:", vol_clean['spot_dir_id'].nunique())

    # 2) Train / Val / Test split
    train_end_v = pd.Timestamp('2024-12-31 23:00:00')
    val_end_v   = pd.Timestamp('2025-06-30 23:00:00')

    vol_train_df = vol_clean[vol_clean['timestamp'] <= train_end_v]
    vol_val_df   = vol_clean[(vol_clean['timestamp'] > train_end_v) & (vol_clean['timestamp'] <= val_end_v)]
    vol_test_df  = vol_clean[vol_clean['timestamp'] > val_end_v]

    vol_stats = vol_clean.groupby("spot_dir_id").agg({
        "도로명": "first",
        "지점명": "first",
        "spot_mean_vol": "first",
        "spot_max_vol": "first",
    }).reset_index()

    vol_meta = {}
    for _, row in vol_stats.iterrows():
        sid = str(row["spot_dir_id"])
        vol_meta[sid] = {
            "road_name": row["도로명"],
            "spot_name": row["지점명"],
            "spot_mean_vol": float(row["spot_mean_vol"]),
            "spot_max_vol": float(row["spot_max_vol"]),
        }

    del vol_clean; gc.collect()

    print("[VOL] train/val/test 행 수:",
          len(vol_train_df), len(vol_val_df), len(vol_test_df))
    print("[VOL] train 기간:", vol_train_df['timestamp'].min(), "~", vol_train_df['timestamp'].max())
    print("[VOL] val   기간:", vol_val_df['timestamp'].min(),   "~", vol_val_df['timestamp'].max())
    print("[VOL] test  기간:", vol_test_df['timestamp'].min(),  "~", vol_test_df['timestamp'].max())

    print("[VOL] train/val/test 내 지점 수:",
          vol_train_df['spot_dir_id'].nunique(),
          vol_val_df['spot_dir_id'].nunique(),
          vol_test_df['spot_dir_id'].nunique())

    # 3) Dataset / DataLoader
    vol_train_ds = VolumeWindowDataset(
        vol_train_df,
        T=T,
        H=H,
        mode='train',
        stride=STRIDE,
    )
    vol_val_ds = VolumeWindowDataset(
        vol_val_df,
        T=T,
        H=H,
        mode='eval',
        spot_id_to_idx=vol_train_ds.spot_id_to_idx,
        scaler_dynamic=vol_train_ds.scaler_dynamic,
        scaler_static=vol_train_ds.scaler_static,
        stride=STRIDE,
    )
    vol_test_ds = VolumeWindowDataset(
        vol_test_df,
        T=T,
        H=H,
        mode='eval',
        spot_id_to_idx=vol_train_ds.spot_id_to_idx,
        scaler_dynamic=vol_train_ds.scaler_dynamic,
        scaler_static=vol_train_ds.scaler_static,
        stride=STRIDE,
    )

    vol_train_loader = DataLoader(
        vol_train_ds, batch_size=BATCH_SIZE, shuffle=True,
        num_workers=0, pin_memory=True if device.type == 'cuda' else False
    )
    vol_val_loader = DataLoader(
        vol_val_ds, batch_size=BATCH_SIZE, shuffle=False,
        num_workers=0, pin_memory=True if device.type == 'cuda' else False
    )
    vol_test_loader = DataLoader(
        vol_test_ds, batch_size=BATCH_SIZE, shuffle=False,
        num_workers=0, pin_memory=True if device.type == 'cuda' else False
    )

    vol_dynamic_dim = vol_train_ds.X_dyn.shape[-1]
    vol_static_dim  = vol_train_ds.X_stat.shape[-1]
    vol_time_dim    = vol_train_ds.future_time.shape[-1]

    X0_vol_tensor, A_hat_vol_tensor = build_spot_graph(vol_train_df, vol_train_ds.spot_id_to_idx)
    gcn_vol_in_dim = X0_vol_tensor.shape[1]  # 2 (spot_mean_vol, spot_max_vol)

    # 4) 모델 초기화
    vol_model = Seq2SeqGCNLSTM(
        dynamic_dim=vol_dynamic_dim,
        static_dim=vol_static_dim,
        horizon=H,
        time_dim=vol_time_dim,
        gcn_in_dim=gcn_vol_in_dim,
        gcn_hidden_dim=GCN_HIDDEN_DIM,
        gcn_out_dim=GCN_OUT_DIM,
        X0=X0_vol_tensor,
        A_hat=A_hat_vol_tensor,
        lstm_hidden_dim=ENC_HIDDEN_DIM,
        lstm_layers=ENC_LAYERS,
        fc_hidden_dim=FC_HIDDEN_DIM,
        dropout_p=DROPOUT_P,
    ).to(device)

    # ✅ 디코더에 타깃 스케일러 주입 (raw>=0 제약을 위해 필수)
    vol_model.decoder.set_target_scaler(vol_train_ds.vol_mean, vol_train_ds.vol_std)

    vol_criterion = nn.SmoothL1Loss()
    vol_optimizer = torch.optim.Adam(vol_model.parameters(), lr=LR)
    vol_scheduler = StepLR(vol_optimizer, step_size=5, gamma=0.5)

    best_vol_val_loss = float('inf')
    vol_train_history, vol_val_history = [], []

    base_weights = [1.0, 1.0, 1.0, 0.8, 0.6, 0.5]
    assert H <= len(base_weights)
    horizon_weights = torch.tensor(base_weights[:H], dtype=torch.float32, device=device)

    def smooth_l1_horizon_weighted(y_hat, y, weights: torch.Tensor):
        loss_per_elem = F.smooth_l1_loss(y_hat, y, reduction='none')  # (B,H)
        weighted = loss_per_elem * weights.unsqueeze(0)
        return weighted.mean()

    # 5) 학습 루프
    for epoch in range(1, EPOCHS+1):
        tf_ratio = max(0.0, 1.0 - (epoch - 1) / (EPOCHS / 2))
        print(f"\n[VOL Epoch {epoch}] Teacher Forcing ratio: {tf_ratio:.3f}")

        # ---- TRAIN ----
        vol_model.train()
        train_loss = 0.0
        train_bar = tqdm(vol_train_loader, desc=f"[VOL Epoch {epoch}] Train", leave=False)

        for x_dyn, spot_idx, x_static, y, last_vol, future_time in train_bar:
            x_dyn = x_dyn.to(device)
            spot_idx = spot_idx.to(device)
            x_static = x_static.to(device)
            y = y.to(device)
            last_vol = last_vol.to(device)
            future_time = future_time.to(device)

            vol_optimizer.zero_grad()

            use_tf = (torch.rand(1).item() < tf_ratio)

            y_hat = vol_model(
                x_dyn, spot_idx, x_static,
                future_time=future_time,
                y=y,
                last_speed=last_vol,
                teacher_forcing=use_tf
            )
            #loss = vol_criterion(y_hat, y)
            loss = smooth_l1_horizon_weighted(y_hat, y, horizon_weights)
            loss.backward()
            vol_optimizer.step()

            batch_loss = loss.item()
            train_loss += batch_loss * x_dyn.size(0)
            train_bar.set_postfix(loss=f"{batch_loss:.4f}", tf=use_tf)

        train_loss /= len(vol_train_ds)

        # ---- VALID ----
        vol_model.eval()
        val_loss = 0.0
        val_bar = tqdm(vol_val_loader, desc=f"[VOL Epoch {epoch}] Valid", leave=False)

        with torch.no_grad():
            for x_dyn, spot_idx, x_static, y, last_vol, future_time in val_bar:
                x_dyn = x_dyn.to(device)
                spot_idx = spot_idx.to(device)
                x_static = x_static.to(device)
                y = y.to(device)
                last_vol = last_vol.to(device)
                future_time = future_time.to(device)

                y_hat = vol_model(
                    x_dyn, spot_idx, x_static,
                    future_time=future_time,
                    y=y,
                    last_speed=last_vol,
                    teacher_forcing=False
                )
                #loss = vol_criterion(y_hat, y)
                loss = smooth_l1_horizon_weighted(y_hat, y, horizon_weights)
                batch_loss = loss.item()
                val_loss += batch_loss * x_dyn.size(0)
                val_bar.set_postfix(loss=f"{batch_loss:.4f}")

        val_loss /= len(vol_val_ds)
        vol_scheduler.step()

        vol_train_history.append(train_loss)
        vol_val_history.append(val_loss)
        print(f"[VOL Epoch {epoch}] train_loss={train_loss:.4f}, val_loss={val_loss:.4f}")

        if val_loss < best_vol_val_loss:
            best_vol_val_loss = val_loss
            torch.save(
                {
                    'model_state_dict': vol_model.state_dict(),
                    'spot_id_to_idx': vol_train_ds.spot_id_to_idx,
                    'scaler_dynamic_state': vol_train_ds.scaler_dynamic.__dict__,
                    'scaler_static_state': vol_train_ds.scaler_static.__dict__,
                    'target_mean': vol_train_ds.vol_mean,
                    'target_std': vol_train_ds.vol_std,
                    'X0': X0_vol_tensor,
                    'A_hat': A_hat_vol_tensor,
                    'config': {
                        'T': T,
                        'H': H,
                        'dynamic_dim': vol_dynamic_dim,
                        'static_dim': vol_static_dim,
                        'time_dim': vol_time_dim,
                        'gcn_in_dim': gcn_vol_in_dim,
                        'gcn_out_dim': GCN_OUT_DIM,
                        'gcn_hidden_dim': GCN_HIDDEN_DIM,
                        'lstm_hidden_dim': ENC_HIDDEN_DIM,
                        'lstm_layers': ENC_LAYERS,
                        'fc_hidden_dim': FC_HIDDEN_DIM,
                        'dropout_p': DROPOUT_P,
                        'lr': LR,
                        'batch_size': BATCH_SIZE,
                        'epochs': EPOCHS,
                        'tf_ratio': tf_ratio,
                        #'initial_tf_ratio': INITIAL_TF_RATIO,
                        #'tf_decay': TF_DECAY,
                        #'min_tf_ratio': MIN_TF_RATIO,
                    },
                    #'optimizer_state_dict': vol_optimizer.state_dict(),
                    #'scheduler_state_dict': vol_scheduler.state_dict(),
                    'vol_train_history': vol_train_history,
                    'vol_val_history': vol_val_history,
                    'best_val_loss': best_vol_val_loss,
                    'vol_meta': vol_meta,
                },
                CKPT_PATH
            )
            print("  -> [VOL] best Seq2Seq volume model updated")

    # 6) 학습 곡선
    plt.figure(figsize=(8, 4))
    plt.plot(range(1, len(vol_train_history)+1), vol_train_history, marker='o', label='vol_train_loss')
    plt.plot(range(1, len(vol_val_history)+1),   vol_val_history,   marker='o', label='vol_val_loss')
    plt.xlabel('Epoch')
    plt.ylabel('SmoothL1 Loss (scaled)')
    plt.title('Volume Seq2Seq GCN-LSTM Train / Validation Loss (scaled target)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 7) best ckpt 로드 후 평가
    ckpt = torch.load(CKPT_PATH, map_location=device, weights_only=False)
    vol_model.load_state_dict(ckpt['model_state_dict'])
    vol_model.to(device)

    #vol_model.decoder.set_target_scaler(ckpt['target_mean'], ckpt['target_std'])

    vol_mean = ckpt['target_mean']
    vol_std  = ckpt['target_std']

    lag24_idx = vol_train_ds.dynamic_cols.index('lag_24h')
    scaler_dyn_state = ckpt['scaler_dynamic_state']
    lag24_mean = float(scaler_dyn_state['mean_'][lag24_idx])
    lag24_std  = float(scaler_dyn_state['scale_'][lag24_idx] + 1e-8)

    model_metrics, (y_true_test, y_pred_test) = evaluate_volume_model_test(
        model=vol_model,
        test_loader=vol_test_loader,
        device=device,
        vol_mean=vol_mean,
        vol_std=vol_std,
        horizon=H
    )

    naive_last_metrics = eval_naive_last_value_volume(
        test_loader=vol_test_loader,
        vol_mean=vol_mean,
        vol_std=vol_std,
        horizon=H
    )

    naive_24h_metrics = eval_naive_last_24h_volume(
        test_loader=vol_test_loader,
        vol_mean=vol_mean,
        vol_std=vol_std,
        horizon=H,
        lag24_idx=lag24_idx,
        lag24_mean=lag24_mean,
        lag24_std=lag24_std
    )


if __name__ == "__main__":
    main()
```
